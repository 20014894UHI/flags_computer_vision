```python
# Step_1_unzipDataExports.jpynb
# Step 1: Files were exported from CVAT in Yolo 1.1. format (supported Yolo format available in CVAT local)
# Unzip zip files exported, keeping the same stem file name. 

import os
import zipfile

def unzip_files(data_to_unzip, unzipped_data):
    # Ensure the output folder exists
    os.makedirs(unzipped_data, exist_ok=True)
    
    # Iterate over all files in the input folder
    for file_name in os.listdir(data_to_unzip):
        # Check if the file is a zip file
        if file_name.endswith('.zip'):
            # Construct the full path to the zip file
            zip_path = os.path.join(data_to_unzip, file_name)
            
            # Create a folder in unzipped_data named after the zip file (without the .zip extension)
            extract_folder_name = os.path.splitext(file_name)[0]
            extract_folder = os.path.join(unzipped_data, os.path.splitext(file_name)[0])
            os.makedirs(extract_folder, exist_ok=True)
            
            # Unzip the file into the new folder
            # with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            #     zip_ref.extractall(extract_folder)
                
            # print(f"Unzipped {file_name} into {extract_folder}")

            try:
                # Unzip the file into the new folder
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    zip_ref.extractall(extract_folder)
                
                print(f"Unzipped {file_name} into {extract_folder}")
                
                # Check if files were extracted
                extracted_files = os.listdir(extract_folder)
                # if not extracted_files:
                #    print(f"Warning: No files extracted for {file_name}.")
                #else:
                #    print(f"Extracted files: {extracted_files}")
                    
            except zipfile.BadZipFile:
                print(f"Error: {file_name} is not a valid zip file.")
            except Exception as e:
                print(f"An error occurred while unzipping {file_name}: {e}")

#Unzip the files 
path_to_data_to_unzip = 'D:\FlagDetectionDatasets\ExportedTaskDatasetsZips\TODO' 
path_to_unzipped_data = 'D:\FlagDetectionDatasets\ExportedDatasetsExtracted'
unzip_files(path_to_data_to_unzip, path_to_unzipped_data)

# Unzip the files to a copy for trimming down for testing 
#path_to_data_to_unzip = 'D:\FlagDetectionDatasets\ExportedTaskDatasetsZips' 
#path_to_unzipped_data = 'D:\FlagDetectionDatasets\ExportedDatasetsNotSplitForTestSet'
#unzip_files(path_to_data_to_unzip, path_to_unzipped_data)
```
