```python
# Move data to a central folder to prepare for splitting into Train, Test, Val 
# Changing the approach. 
import os
import shutil

parent_dir = "D:/FlagDetectionDatasets/ExportedDatasetsSelectedML"
central_dir = "D:/FlagDetectionDatasets/ExportedDatasetsSelectedMLSTARTSET"

#if not os.path.exists(central_dir):
#    os.makedirs(central_dir)

# List to track files that were not moved
unmoved_files = []
# Dictionary to track seen filenames (for checking duplicates)
seen_files = {}

# Move images and annotations to the central folder
for root, dirs, files in os.walk(parent_dir, topdown=False):  # topdown=False ensures we handle empty folders last
    for filename in files:
        # Only move image and annotation pairs
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')) or filename.lower().endswith('.txt'):
            # If it's an image, find its corresponding annotation file
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                annotation_file = filename.rsplit('.', 1)[0] + '.txt'
                annotation_path = os.path.join(root, annotation_file)

                # If there's a matching annotation file, move the pair
                if os.path.exists(annotation_path):
                    img_path = os.path.join(root, filename)

                    # Check for duplicate filenames and skip if necessary
                    if filename in seen_files:
                        # If duplicate, skip and don't move it
                        continue
                    else:
                        seen_files[filename] = 0

                    # Move files
                    dest_img_path = os.path.join(central_dir, filename)
                    dest_ann_path = os.path.join(central_dir, annotation_file)

                    shutil.move(img_path, dest_img_path)
                    shutil.move(annotation_path, dest_ann_path)
                else:
                    # If annotation file doesn't exist, add image to unmoved list
                    unmoved_files.append(filename)
            else:
                # If it's a text file, check if it has a corresponding image
                image_file = filename.rsplit('.', 1)[0] + '.png'  # Assume image is in PNG format
                if not os.path.exists(os.path.join(root, image_file)):
                    # If image file doesn't exist, add annotation to unmoved list
                    unmoved_files.append(filename)

    # After processing files, attempt to remove empty directories
    if not os.listdir(root):  # Check if the directory is empty
        os.rmdir(root)

# Print list of unmoved files (if any)
if unmoved_files:
    print("The following files were not moved (no matching pair found):")
    for file in unmoved_files:
        print(file)
else:
    print("All files were successfully moved.")

#print("File move and cleanup completed.")
# Print timestamp of completion
completion_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
print(f"File move and cleanup completed at {completion_time}.")
```
