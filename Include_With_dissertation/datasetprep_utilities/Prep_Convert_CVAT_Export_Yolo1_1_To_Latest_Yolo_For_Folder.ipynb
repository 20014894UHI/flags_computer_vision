```python
# Listing - rename files in the ob_train_data sub-folder - cvat exports in this structure but we want to reconfigure it 
# rename all files that have 'frame' in the title - replace 'frame' with the job name 
# move all the files out of the obj_tain_data folder and up a level 
# delete the obj_train folder and other file remnants

# Listing - Rename and reorganize files exported from CVAT - SPECIFIC FOLDER VERSION 
# Clean up and restructure files in a folder named obj_train_data within a job folder. 
# - construct the path to the obj_train_data folder within the provided job_folder.
# - iterate over all files in the obj_train_data folder
# - replace 'frame' with the jobNo provided (e.g., 'Job_13')
# - Move renamed files from the obj_train_data folder to its parent folder (job_folder).
# - Clean Up the obj_train_data Folder: delete the empty obj_train_data folder.
# - Remove unnecessary Files: check for and delete three specific files from the job_folder: obj.data, obj.names, train.txt
# - Print status messages for renamed files, folder removals, and file deletions.

import os
import shutil

def rename_and_move_files(job_folder, jobNo):
    # Define the folder from where files need to be moved
    obj_train_data_path = os.path.join(job_folder, "obj_train_data")
    
    # Check if the obj_train_data folder exists
    if not os.path.exists(obj_train_data_path):
        print(f"The path {obj_train_data_path} does not exist.")
        return
    
    # Iterate over all files in the obj_train_data folder
    for filename in os.listdir(obj_train_data_path):
        file_path = os.path.join(obj_train_data_path, filename)

        # Check if it's a file (not a folder)
        if os.path.isfile(file_path):
            # Custom renaming logic (example: replacing 'frame' with 'Job_13')
            new_filename = filename.replace('frame', jobNo)
            # Define the new path for the renamed file in the parent folder
            new_file_path = os.path.join(job_folder, new_filename)
            # Rename and move the file to the parent folder
            shutil.move(file_path, new_file_path)
            # print(f"Moved and renamed: {file_path} -> {new_file_path}")
    
    # Remove the empty obj_train_data folder after moving all files
    if not os.listdir(obj_train_data_path):
        os.rmdir(obj_train_data_path)
        print(f"Removed empty folder: {obj_train_data_path}")
    else:
        print(f"Folder {obj_train_data_path} is not empty, cannot remove.")

    # Remove obj.data and obj.names if they exist in the job_folder
    data_file = os.path.join(job_folder, "obj.data")
    names_file = os.path.join(job_folder, "obj.names")
    old_text_file = os.path.join(job_folder, "train.txt")

    if os.path.exists(data_file):
        os.remove(data_file)
        print(f"Removed file: {data_file}")

    if os.path.exists(names_file):
        os.remove(names_file)
        print(f"Removed file: {names_file}")

    if os.path.exists(old_text_file):
        os.remove(names_file)
        print(f"Removed file: {old_text_file}")
        
    print(f"Done")


def rename_and_move_files_flat(job_folder, jobNo):
    
    # Iterate over all files in the obj_train_data folder
    for filename in os.listdir(job_folder):
        file_path = os.path.join(job_folder, filename)

        # Check if it's a file (not a folder)
        if os.path.isfile(file_path):
            # Custom renaming logic (example: replacing 'frame' with 'Job_13')
            new_filename = filename.replace('frame', jobNo)
            # Define the new path for the renamed file in the parent folder
            new_file_path = os.path.join(job_folder, new_filename)
            # Rename and move the file to the parent folder
            shutil.move(file_path, new_file_path)
            # print(f"Moved and renamed: {file_path} -> {new_file_path}")
    
    # Remove obj.data and obj.names if they exist in the job_folder
    data_file = os.path.join(job_folder, "obj.data")
    names_file = os.path.join(job_folder, "obj.names")
    old_text_file = os.path.join(job_folder, "train.txt")

    if os.path.exists(data_file):
        os.remove(data_file)
        print(f"Removed file: {data_file}")

    if os.path.exists(names_file):
        os.remove(names_file)
        print(f"Removed file: {names_file}")

    if os.path.exists(old_text_file):
        os.remove(names_file)
        print(f"Removed file: {old_text_file}")
        
    print(f"Done")

# Call the function to clean, rename and reorganise from CVAT 1.1. TO Yolo
#job = "Job_36"
job = "Job_118"
folder_path = r"D:\FlagDetectionDatasets\ExportedDatasetsReduced"  
#folder_path = r"D:\FlagDetectionDatasets\Augmentation\scaled\augmented"
job_folder = os.path.join(folder_path, job)
#job_folder = r"D:\FlagDetectionDatasets\ExportedDatasetsNotSplit\Job13"  
#rename_and_move_files(job_folder, job)
rename_and_move_files_flat(job_folder, job)

#Removed empty folder: D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_36\obj_train_data
#Removed file: D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_36\obj.data
#Removed file: D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_36\obj.names
```
