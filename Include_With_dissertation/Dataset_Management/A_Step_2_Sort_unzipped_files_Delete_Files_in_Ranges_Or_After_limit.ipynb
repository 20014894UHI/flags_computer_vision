```python
# A_Step_2_Sort_unzipped_files_Delete_Files_in_Ranges_Or_After_limit.ipynb
# latest - arrange unzipped files - filter some out to a folder for testing 
# Data cleansing and pre-processing to move files that are not to be used for training 
# Move files and ranges of file from the initial files unzipped. 

# Example output: 
# Deleting files in range 1901 to 5000 for Job_36...
# Completed deletion of files in range 1901 to 5000 for Job_36.
# Remaining pairs after deletion: 1901

# Starting Job_27...
# No files to delete. Total images in Job_27: 161
# Starting Job_28...
# No files to delete. Total images in Job_28: 2089
# Starting Job_37...
# Error: The folder D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_37\obj_train_data does not exist.
# Deleting files in range 185 to 3000 for Job_52...
# Completed deletion of files in range 185 to 3000 for Job_52.
# Remaining pairs after deletion: 132

import os
import shutil

base_path_src = r'D:\FlagDetectionDatasets\ExportedDatasetsExtracted'
base_path_tgt = r'D:\FlagDetectionDatasets\ExportedDatasetsReduced'
#base_path_src = r'D:\FlagDetectionDatasets\ExportedDatasetsNotSplit'
#base_path_tgt = r'D:\FlagDetectionDatasets\ExportedDatasetsExtractedForTestSet'
sub_folder_train = r'obj_train_data'
sub_folder_test = r'obj_train_data'

# 1. Copy FOlders where not extracting a range 
def copy_folders(folder_paths, destination_folder):
    # Ensure the destination folder exists
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for folder_path in folder_paths:

        folder_path  = os.path.join(base_path_src, folder_path)
        #print(folder_path)
        #path_tgt =  os.path.join(base_path_tgt, folder_)
        #path_tgt  = os.path.join(path_tgt, sub_folder)

        if os.path.isdir(folder_path):  # Check if the path is a directory
            try:
                folder_name = os.path.basename(folder_path)
                dest_path = os.path.join(destination_folder, folder_name)
                # Move the folder to the destination
                shutil.copy(folder_path, dest_path)
                #shutil.move(folder_path, dest_path)
                print(f"Copied folder '{folder_path}' to '{dest_path}'")
            except Exception as e:
                print(f"Error copying folder '{folder_path}': {e}")
        else:
            print(f"Path is not a folder or doesn't exist: {folder_path}")

# 1. Copy folders 
# Done 
# folders_to_copy = [ 'Job_7', 'Job_8', 'Job_11', 'Job_12', 'Job_13', 'Job_14', 'Job_15', 'Job_16', 'Job_17', 'Job_18']
# folders_to_copy = [ 'Job_21','Job_11','Job_22']
# copy_folders(folders_to_copy, base_path_tgt)

# 2. Dataset preparation  - DELETE  a range of files from the unzipped files initial set 
# Traverse the Yolo 1.1. folder structure 
## Delete files after an upper limit when ordered by name from a folder and count remaining pairs 
def delete_files_after_limit(whichfolder, folder, limit):
    """
    Deletes image and annotation files after the first `limit` files when ordered by name.
    Args:
        whichfolder (str): The base directory containing the folders.
        folder (str): The specific folder to process.
        sub_folder_train (str): The sub-folder inside the folder to process.
        limit (int): Number of files to keep, after which files will be deleted.
    """
    print(f"Starting {folder}...")

    # Build the path to the target folder
    path = os.path.join(whichfolder, folder, sub_folder_train)

    # Ensure the folder exists
    if not os.path.exists(path):
        print(f"Error: The folder {path} does not exist.")
        return

    # List all files in the folder
    files = os.listdir(path)

    # Filter and sort files by name
    image_files = sorted([f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    #image_files = sorted([f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    text_files = sorted([f for f in files if f.lower().endswith('.txt')])

    # Ensure file pairs (images and corresponding annotations)
    paired_files = []
    for image_file in image_files:
        base_name = os.path.splitext(image_file)[0]
        annotation_file = f"{base_name}.txt"
        if annotation_file in text_files:
            paired_files.append((image_file, annotation_file))

    # Check if there are more than the limit
    if len(paired_files) <= limit:
        print(f"No files to delete. Total images in {folder}: {len(paired_files)}")
        return

    # Files to delete (after the first `limit` pairs)
    files_to_delete = paired_files[limit:]

    # Delete the files
    for image_file, annotation_file in files_to_delete:
        try:
            # Delete the image file
            os.remove(os.path.join(path, image_file))
            # Delete the annotation file
            os.remove(os.path.join(path, annotation_file))
        except Exception as e:
            print(f"Error deleting files {image_file} or {annotation_file}: {e}")

    print(f"Deleted {len(files_to_delete)} file pairs from {path}")

## 2. Better to read from the Excel if have to redo ...
## Copy 21 to reduced then unzip again 

#delete_files_after_limit(base_path_tgt, folder="Job_108",   limit=50)
#delete_files_after_limit(base_path_tgt, folder="Job_116",   limit=300)

#delete_files_after_limit(base_path_tgt, folder="Job_32",   limit=4000)

#delete_files_after_limit(base_path_tgt, folder="Job_21",   limit=221)
#delete_files_after_limit(base_path_tgt, folder="Job_22",   limit=2000) # Delete the files after one named 2000
#delete_files_after_limit(base_path_tgt, folder="Job_31",   limit=2500)

# Note: Filtered out non-ovelapping flags first for 24
#delete_files_after_limit(base_path_tgt, folder="Job_24",   limit=1100)
#delete_files_after_limit(base_path_tgt, folder="Job_28",   limit=2501)


# ************************* CHECK 11  EXTRACT 11 AND 21 AGAIN  

#delete_files_after_limit(base_path_tgt, folder="Job_32",   limit=2500)
#delete_files_after_limit(base_path_tgt, folder="Job_41",   limit=1400)

#delete_files_after_limit(base_path_tgt, folder="Job_50",   limit=1200)
#delete_files_after_limit(base_path_tgt, folder="Job_52",   limit=184)
#delete_files_after_limit(base_path_tgt, folder="Job_59",   limit=348)
#delete_files_after_limit(base_path_tgt, folder="Job_61",   limit=550)
#delete_files_after_limit(base_path_tgt, folder="Job_70",   limit=600)

# STEP 2 
#delete_files_after_limit(base_path_tgt, folder="Job_27",   limit=1000)
#delete_files_after_limit(base_path_tgt, folder="Job_28",   limit=2120)
#delete_files_after_limit(base_path_tgt, folder="Job_37",   limit=2003)

#delete_files_after_limit(base_path_tgt, folder="Job_29",   limit=1200)
#delete_files_after_limit(base_path_tgt, folder="Job_30",   limit=2500)
#delete_files_after_limit(base_path_tgt, folder="Job_31",   limit=2500)

# NEXT ***UNZIP afterexport AND COPY 

#Deleted 2204 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_30\obj_train_data
#Deleted 2312 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_31\obj_train_data
#Deleted 812 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_32\obj_train_data
#Deleted 1261 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_37\obj_train_data
#Deleted 172 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_41\obj_train_data

#Deleted 708 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_50\obj_train_data
#Deleted 1257 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_52\obj_train_data
#Deleted 192 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_59\obj_train_data
#Deleted 1286 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_61\obj_train_data
#Deleted 1092 file pairs from D:\FlagDetectionDatasets\ExportedDatasetsReduced\Job_70\obj_train_data


## 3DUD. Delete a range of files corresponding to filenames passed in (not the index of the file)

def delete_files_in_range(base_folder, job, min_number=0, max_number=54):
    """
    Deletes image/annotation pairs from a folder structure based on filename range.
    :param base_folder: Path to the folder containing job folders.
    :param job: Name of the job folder to process.
    :param min_number: Minimum frame number to delete.
    :param max_number: Maximum frame number to delete.
    """
    print(f"Deleting files in range {min_number} to {max_number} for {job}...")

    job_path = os.path.join(base_folder, job)
    obj_train_data_path = job_path #os.path.join(job_path, "obj_train_data")

    #if not os.path.exists(obj_train_data_path):
    #    print(f"No obj_train_data folder in {job}, skipping...")
    #    return

    # Loop through sorted files in the folder
    for file in sorted(os.listdir(obj_train_data_path)):
        # Get file number from filename (assuming the format is frame_000000.ext)
        filename, ext = os.path.splitext(file)
        if ext.lower() not in ['.png', '.jpg', '.jpeg', '.txt']:
            continue  # Skip non-image/annotation files

        # Extract the frame number from the filename
        try:
            frame_number = int(filename.split('_')[-1])  # Assumes format frame_000000
        except ValueError:
            #print(f"Skipping file with unexpected format: {file}")
            continue

        # Check if the frame number is in the specified range
        if min_number <= frame_number <= max_number:
            # Construct full path and delete file
            file_path = os.path.join(obj_train_data_path, file)
            os.remove(file_path)
            #print(f"Deleted: {file_path}")

    print(f"Completed deletion of files in range {min_number} to {max_number} for {job}.")
    # Print the completion message
    # Recalculate remaining pairs after deletion
    remaining_files = os.listdir(obj_train_data_path)
    remaining_image_files = sorted([f for f in remaining_files if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    remaining_text_files = sorted([f for f in remaining_files if f.lower().endswith('.txt')])
    remaining_paired_files = [
        (image_file, f"{os.path.splitext(image_file)[0]}.txt")
        for image_file in remaining_image_files
        if f"{os.path.splitext(image_file)[0]}.txt" in remaining_text_files
    ]
    print(f"Remaining pairs after deletion: {len(remaining_paired_files)}")


if __name__ == "__main__":
    #delete_files_in_range(base_path_tgt, job="Job_67", min_number=0, max_number=54)
    # Job_69 - use "0-1480, (onl a few), 3348-3627,  3849-3910,  4180-4230; "
    #delete_files_in_range(base_path_tgt, job="Job_31", min_number=2201, max_number=4800)  
    #delete_files_in_range(base_path_tgt, job="Job_32", min_number=2501, max_number=7000)  

    #delete_files_in_range(base_path_tgt, job="Job_31", min_number=, max_number=5000)  
    #delete_files_in_range(base_path_tgt, job="Job_27", min_number=1801, max_number=6000)  # CVAT EXPORT STARTS AT 500 
    # So for first 1300 = filenames 500 + to 1800 
    # delete_files_in_range(base_path_tgt, job="Job_52", min_number=185, max_number=3000)  
    #delete_files_in_range(base_path_tgt, job="Job_69", min_number=3909, max_number=4179)
    #delete_files_in_range(base_path_tgt, job="Job_69", min_number=1481, max_number=3347)
    #delete_files_in_range(base_path_tgt, job="Job_69", min_number=3628, max_number=3848)
    #delete_files_in_range(base_path_tgt, job="Job_69", min_number=4231, max_number=5500)
    # 66 
    #delete_files_in_range(base_path_tgt,    folder="Job_66",min_number=181,max_number=1499)
    # 67:Use 55-62 1152-1173 1264 - 1280. 1321-22, 1480-1512. 1604 - 1607. 
    #delete_files_in_range(base_path_tgt,    folder="Job_67",min_number=1513,max_number=1603)
    #delete_files_in_range(base_path_tgt,    folder="Job_67",min_number=1323,max_number=1479)
    #delete_files_in_range(base_path_tgt,    folder="Job_67",min_number=1281,max_number=1320)
    #delete_files_in_range(base_path_tgt,    folder="Job_67",min_number=1174,max_number=1263)
    #delete_files_in_range(base_path_tgt,    folder="Job_67",min_number=63,max_number=1151)
    #delete_files_in_range(base_path_tgt,    folder="Job_67",min_number=0,max_number=54)


# OLDdelete_files_in_range(base_path_tgt,    folder="Job_21",min_number=221,max_number=1500)

```
