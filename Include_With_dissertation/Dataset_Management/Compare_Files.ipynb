```python
import os

def compare_folders(folder1, folder2):
    print(folder1)

    differences = []

    # Get all files from folder1
    for root, _, files in os.walk(folder1):
        for file in files:
            print(f"Comparing {file}")

            file1_path = os.path.join(root, file)
            file1_path = os.path.normpath(file1_path)
            print(file1_path)
            # Look for the exact matching file in folder2
            for sub_root, _, sub_files in os.walk(folder2):
                #print(sub_files)
                if file in sub_files:
                    file2_path = os.path.join(sub_root, file)
                    file2_path = os.path.normpath(file2_path)
                    print(f"TO {file2_path}/n")
                    # Compare file content line by line
                    with open(file1_path, "r") as f1, open(file2_path, "r") as f2:
                        file1_lines = f1.readlines()
                        file2_lines = f2.readlines()
                        #print(file1_lines)
                    differences_in_content = []
                    for i, (line1, line2) in enumerate(zip(file1_lines, file2_lines)):
                        print(line1)
                        print(line2)
                        if line1 != line2:
                            differences_in_content.append(
                                {
                                    "line_number": i + 1,
                                    "file1_line": line1.strip(),
                                    "file2_line": line2.strip(),
                                }
                            )

                    # Record differences if any mismatches found
                    if differences_in_content:
                        differences.append(
                            {
                                "file": file,
                                "folder1_path": file1_path,
                                "folder2_path": file2_path,
                                "differences": differences_in_content,
                            }
                        )

    return differences

#folder1 = "C:/Users/peter/projects/classifyflags/Include_With_dissertation/datasetprep_utilities"
folder1 = 'C:/Users/peter/projects/classifyflags/Include_With_dissertation/Augmentation/.ipynb_checkpoints'
folder2='D:/classifyflags_after_restoring_from_checkpoint/Include_With_dissertation/Augmentation/.ipynb_checkpoints'
differences = compare_folders(folder1, folder2)
# Display content differences
print("\nContent differences:")
for diff in differences:
    print(f"{diff['file']}\n  Folder1: {diff['folder1_path']}\n  Folder2: {diff['folder2_path']}")
    for content_diff in diff["differences"]:
          print(f"    Line {content_diff['line_number']}\n      Folder1: {content_diff['file1_line']}\n      Folder2: {content_diff['file2_line']}")
```
