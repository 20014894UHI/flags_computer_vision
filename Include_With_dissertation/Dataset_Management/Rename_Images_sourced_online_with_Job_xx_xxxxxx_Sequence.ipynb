```python
# Listing renameImagesWith_Job_xx_xxxxxx_Sequence.jpynb
# For tidying up / standardising image/annotation pairs filenames and renaming sets of downladoed images that had random filenames 
# These output examples demonstrate what it does 
# The script was used to tidy up/ standardize the nameing schema of images sources online
# Renamed 'img-1186-800x600.jpg' to 'Job_154_000001.jpg'
# Renamed annotation 'img-1186-800x600.txt' to 'Job_154_000001.txt'
# Renamed '107-12767-400.jpg' to 'Job_154_000002.jpg'
# Renamed annotation '107-12767-400.txt' to 'Job_154_000002.txt'
# Renamed '32106040792_edca5fbef3_k.jpg' to 'Job_154_000003.jpg'
# Renamed annotation '32106040792_edca5fbef3_k.txt' to 'Job_154_000003.txt'
# Renamed 'blue_flag_beaches_3.jpg' to 'Job_154_000004.jpg'
# Renamed annotation 'blue_flag_beaches_3.txt' to 'Job_154_000004.txt'
# Renamed 'red-yellow-warning-flag-beach-ballybunion-ireland-wild-atlantic-way-72054744.webp' to 'Job_154_000005.webp'
# Renamed annotation 'red-yellow-warning-flag-beach-ballybunion-ireland-wild-atlantic-way-72054744.txt' to 'Job_154_000005.txt'

import os

def rename_images(folder_path, job_number):
    # Get all files in the folder
    files = sorted([f for f in os.listdir(folder_path) if f.endswith.lower((('.png', '.jpg', '.jpeg', 'webp')))])
    
    # Rename each image file
    for index, filename in enumerate(files, start=1):
        # Get file extension
        file_extension = os.path.splitext(filename)[1]
        
        # Create new file name
        #new_filename = f"Job_{job_number}_{index}{file_extension}"
        new_filename = f"{job_number}_{index}{file_extension}"
        
        # Get full path for current and new file
        current_file_path = os.path.join(folder_path, filename)
        new_file_path = os.path.join(folder_path, new_filename)
        
        # Rename the file
        os.rename(current_file_path, new_file_path)
        print(f"Renamed {filename} to {new_filename}")

## REMAMING FOR DOWNLOADED IMAGES 
import os

def rename_files_in_folder(folder_path, prefix, start_number=1, padding=6):
    # Ensure the folder path exists
    if not os.path.exists(folder_path):
        print(f"The folder '{folder_path}' does not exist.")
        return
    
    # Get all files in the folder
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    
    # Separate image and text files
    image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg', '.webp'))]
    
    # Loop through each image file and rename it
    for i, image_filename in enumerate(image_files, start=start_number):
        # Get the file extension and base name
        file_extension = os.path.splitext(image_filename)[1].lower()
        file_base = os.path.splitext(image_filename)[0]
        
        # Format the sequence number with zero padding
        sequence_number = str(i).zfill(padding)
        
        # Construct the new filename for the image
        new_image_filename = f"{prefix}{sequence_number}{file_extension}"
        
        # Full path to the old and new image file names
        old_image_path = os.path.join(folder_path, image_filename)
        new_image_path = os.path.join(folder_path, new_image_filename)
        
        # Rename the image file
        os.rename(old_image_path, new_image_path)
        print(f"Renamed '{image_filename}' to '{new_image_filename}'")

        # Check for corresponding .txt annotation file
        old_annotation_filename = f"{file_base}.txt"
        old_annotation_path = os.path.join(folder_path, old_annotation_filename)
        
        if os.path.exists(old_annotation_path):
            # Create the new annotation filename and rename it
            new_annotation_filename = f"{prefix}{sequence_number}.txt"
            new_annotation_path = os.path.join(folder_path, new_annotation_filename)
            os.rename(old_annotation_path, new_annotation_path)
            print(f"Renamed annotation '{old_annotation_filename}' to '{new_annotation_filename}'")
        else:
            print(f"No annotation found for '{image_filename}', skipping.")

# folder_path = "D:/FlagDetectionDatasets/ONline_datasets_in_progress/DL_Other_beach_flags"
# rename_files_in_folder(folder_path, "OBF_", start_number=1, padding=6)

# folder_path = "D:/FlagDetectionDatasets/ONline_datasets_in_progress/DL_Other_beach_flags"
# rename_files_in_folder(folder_path, "OBF_", start_number=1, padding=6)

# RENAME THE OTHER BEACH FLAGS IMAGES TO A NUMERIC SEQUENCE 
#folder_path = "D:/FlagDetectionDatasets/ONline_datasets_in_progress/DL_Other_beach_flags"
#rename_files_in_folder(folder_path, "OBF_", start_number=1, padding=6)

# RENAME THE BLUE FLAGS AND GREEN COAST FLAGS DOWNLOADED AND MODIFIED IMAGES TO NUMERIC SEQUENCE 
#folder_path ="D:/FlagDetectionDatasets/ONline_datasets_in_progress/DL_BLUE_GREEN"
#rename_files_in_folder(folder_path, "BG_", start_number=1, padding=6)

# RENAME THE DOWNLADED AND MODIFIED REDFLAGIMAGES TO A NUMERIC SEQUENCE 
#folder_path ="D:/FlagDetectionDatasets/ONline_datasets_in_progress/DL_RED"
#rename_files_in_folder(folder_path, "RED_", start_number=1, padding=6)

# RENAME THE DOWNLADED AND MODIFIED RED OVER YELLOW  TO A NUMERIC SEQUENCE 
#folder_path ="D:/FlagDetectionDatasets/ONline_datasets_in_progress/DL_Red_Yellow"
#rename_files_in_folder(folder_path, "RYF_", start_number=1, padding=6)

# oTHER BEACH ARTEFACTS 
#folder_path ="D:/FlagDetectionDatasets/ONline_datasets_in_progress/DL_other_beach_objects"
#rename_files_in_folder(folder_path, "OBO_", start_number=1, padding=6)

folder_path = "D:\FlagDetectionDatasets\ExportedDatasetsExtracted\Job_154\obj_train_data"
rename_files_in_folder(folder_path, "Job_154_", start_number=1, padding=6)

#folder_path = 'D:/FlagDetectionDatasets/A_downloaded_bgs_1280_720'
#job_number = 123  
#rename_images(folder_path, 'D2')
#rename_images(folder_path, 148)

# Example output 
#Renamed 'img-1186-800x600.jpg' to 'Job_154_000001.jpg'
#Renamed annotation 'img-1186-800x600.txt' to 'Job_154_000001.txt'
#Renamed '107-12767-400.jpg' to 'Job_154_000002.jpg'
#Renamed annotation '107-12767-400.txt' to 'Job_154_000002.txt'
#Renamed '32106040792_edca5fbef3_k.jpg' to 'Job_154_000003.jpg'
#Renamed annotation '32106040792_edca5fbef3_k.txt' to 'Job_154_000003.txt'
#Renamed 'blue_flag_beaches_3.jpg' to 'Job_154_000004.jpg'
#Renamed annotation 'blue_flag_beaches_3.txt' to 'Job_154_000004.txt'
#Renamed 'red-yellow-warning-flag-beach-ballybunion-ireland-wild-atlantic-way-72054744.webp' to 'Job_154_000005.webp'
#Renamed annotation 'red-yellow-warning-flag-beach-ballybunion-ireland-wild-atlantic-way-72054744.txt' to 'Job_154_000005.txt'
#Renamed 'life-saving-flag-2802437.webp' to 'Job_154_000006.webp'
#Renamed annotation 'life-saving-flag-2802437.txt' to 'Job_154_000006.txt'
#Renamed 'frame_000007.png' to 'Job_154_000007.png'
#Renamed annotation 'frame_000007.txt' to 'Job_154_000007.txt'
#Renamed '45.png' to 'Job_154_000008.png'
#Renamed annotation '45.txt' to 'Job_154_000008.txt'
#Renamed 'istockphoto-458347231-612x612.jpg' to 'Job_154_000009.jpg'
#Renamed annotation 'istockphoto-458347231-612x612.txt' to 'Job_154_000009.txt'
#Renamed 'train.png' to 'Job_154_000010.png'
#Renamed annotation 'train.txt' to 'Job_154_000010.txt'
#Renamed '3150128432_92729f4b5a_o.png' to 'Job_154_000011.png'
#Renamed annotation '3150128432_92729f4b5a_o.txt' to 'Job_154_000011.txt'
#Renamed 'galazies.jpg' to 'Job_154_000012.jpg'
#Renamed annotation 'galazies.txt' to 'Job_154_000012.txt'
#Renamed 'frame_000022.png' to 'Job_154_000013.png'
#Renamed annotation 'frame_000022.txt' to 'Job_154_000013.txt'
#Renamed 'blue-flag-1504428.jpg' to 'Job_154_000014.jpg'
#Renamed annotation 'blue-flag-1504428.txt' to 'Job_154_000014.txt'
```
