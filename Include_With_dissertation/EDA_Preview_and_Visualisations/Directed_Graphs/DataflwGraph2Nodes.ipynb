```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(format='png', engine='dot')

# Set global attributes for better visualization
dot.attr(dpi='300', rankdir='TB')  # Vertical layout (top-to-bottom)
dot.node('A', 'Extracted set [Approx 100k images]', shape='box', style='filled', fillcolor='lightblue')

# Add the three new boxes
dot.node('B', 'Job_23 in "Reduced" set \n High quality set \n [2460 images/2GB]', shape='box', style='filled', fillcolor='lightyellow')
dot.node('C_51', 'Job_51 in "Reduced" set \n "Green Coast flag coexists with \n incorrectly hanging red/yellow flag \n [1595 images/1.3GB]', shape='box', style='filled', fillcolor='lightyellow')
dot.node('D', 'Other jobs in "Reduced" set \n', shape='box', style='filled', fillcolor='lightyellow')

# Add a background rectangle slightly offset to the left and top of node D
#with dot.subgraph() as s:
#    s.attr(style='dashed', color='black')  # Set style for the background rectangle
#    s.node('D_background', '', shape='rect', width='1.5', height='1', color='transparent', style='dashed')
#dot.node('Invisible', '', width='0', height='0', style='invisible')
#dot.edge('Invisible', 'D', style='invisible')

# Arrows with labels from "Extracted" to each of the jobs
dot.edge('A', 'B', label='', fontsize='14', fontcolor='black')
dot.edge('A', 'C_51', label='Copy "Extracted" set to create "Reduced" set', fontsize='14', fontcolor='black')
dot.edge('A', 'D', label='', fontsize='14', fontcolor='black')


dot.node('Circle', '', shape='ellipse', width='0.1', height='0.1', style='filled', fillcolor='white')
dot.edge('B', 'Circle', label='Separate flags based on overlap of Blue flag with Red flag', fontsize='14', fontcolor='black')

# Add two new boxes below the circle, side by side
dot.node('B1', 'Job_23 \n[2200 images]', shape='box', style='filled', fillcolor='lightgreen')
dot.node('B2', 'Job_23_filter \n[260 images]', shape='box', style='filled', fillcolor='lightpink')

# Ensure B1 and B2 are on the same rank
with dot.subgraph() as s:
    s.attr(rank='same')
    s.node('B1')
    s.node('B2')

dot.edge('Circle', 'B1', label='Blue flag overlaps red flag', fontsize='14', fontcolor='black')
dot.edge('Circle', 'B2', label='Blue flag does not overlap red flag', fontsize='14', fontcolor='black')
dot.node('B1a', 'Job_23 \n [220 Images]', shape='box', style='filled', fillcolor='lightblue')
dot.edge('B1', 'B1a', label='Keep 1 in 10 pattern', fontsize='14', fontcolor='black')
dot.node('Selected23', 'Job_23 \n added to "Selected" set', shape='box', style='filled', fillcolor='lightgrey')
dot.node('Selected23filter', 'Job_23_filter \n added to "Selected" set', shape='box', style='filled', fillcolor='lightgrey')

# Ensure Selected23 and Selected23filter are on the same rank (side by side)
with dot.subgraph() as s:
    s.attr(rank='same')
    s.node('Selected23')
    s.node('Selected23filter')

dot.edge('B1a', 'Selected23', label='Copy to "Selected" Set', fontsize='14', fontcolor='black')
dot.edge('B2', 'Selected23filter', label='Copy to "Selected" Set', fontsize='14', fontcolor='black')

dot.node('S1', 'Job_23 \n (Reduced by pattern) \n[148 images]', shape='box', style='filled', fillcolor='lightcoral')
dot.node('S2', 'Job_23_filtered \n (Keep 1 in 3) \n[85 images]', shape='box', style='filled', fillcolor='lightcoral')

dot.edge('Selected23', 'S1', label='Analyse images/quantity/quality \n -> Reduce by pattern', fontsize='14', fontcolor='black')
dot.edge('Selected23filter', 'S2', label='', fontsize='14', fontcolor='black')

dot.node('S1aug', 'Job_23_Aug \n [148 images]', shape='box', style='filled', fillcolor='lightgreen')
dot.node('S1orig', 'Job_23 \n [148 images]', shape='box', style='filled', fillcolor='lightgreen')

dot.edge('S1', 'S1aug', label='Augmentation: Scale', fontsize='14', fontcolor='black')
dot.edge('S1', 'S1orig', label=' ', fontsize='14', fontcolor='black')

dot.node('S23AugSpliced', 'Job_23_Aug \n [74 images]', shape='box', style='filled', fillcolor='lightblue')

dot.edge('S1aug', 'S23AugSpliced', label='Splice folders to delete alternate pairs', fontsize='14', fontcolor='black')

dot.node('S23Spliced', 'Job_23 \n [75 images]', shape='box', style='filled', fillcolor='lightblue')
dot.edge('S1orig', 'S23Spliced', label='', fontsize='14', fontcolor='black')
dot.node('S23Merged', 'Merged set \n [147 images, 129MB]', shape='box', style='filled', fillcolor='lightyellow')

dot.edge('S23AugSpliced', 'S23Merged', label='', fontsize='14', fontcolor='black')
dot.edge('S23Spliced', 'S23Merged', label='', fontsize='14', fontcolor='black')

# --- Add full sequence for S2 ---
dot.node('S2aug', 'Job_23_Filter_Aug \n [85 images]', shape='box', style='filled', fillcolor='lightgreen')
dot.edge('S2', 'S2aug', label='Augmentation: Scale', fontsize='14', fontcolor='black')

dot.node('S223F', 'Job_23_Filter \n [85 images]', shape='box', style='filled', fillcolor='lightgreen')
dot.edge('S2', 'S223F', label='', fontsize='14', fontcolor='black')

dot.node('S23FilterAugSpliced', 'Job_23_Filter_Aug \n Spliced \n [42 images]', shape='box', style='filled', fillcolor='lightblue')
dot.edge('S2aug', 'S23FilterAugSpliced', label='Splice folders to delete alternate pairs', fontsize='14', fontcolor='black')

dot.node('S23FilterSpliced', 'Job_23_Filter \n [43 images]', shape='box', style='filled', fillcolor='lightblue')
dot.edge('S223F', 'S23FilterSpliced', label='', fontsize='14', fontcolor='black')

dot.node('S23FilterMerged', 'Merged set \n [85 images, 75MB]', shape='box', style='filled', fillcolor='lightyellow')
dot.edge('S23FilterAugSpliced', 'S23FilterMerged', label='', fontsize='14', fontcolor='black')
dot.edge('S23FilterSpliced', 'S23FilterMerged', label='', fontsize='14', fontcolor='black')

# Add the new box and arrow for illumination
dot.node('C_51_illum', 'Folder: Job_51 Illum \n [1595 images]', shape='box', style='filled', fillcolor='lightyellow')
dot.edge('C_51', 'C_51_illum', label="Images underexposed. \n Improve illumination \n Optimise for hsv related processing", fontsize='14', fontcolor='black')

dot.node('C_51_illumA', 'Folder: Job_51 Illum \n Green Cost with partial red flag', shape='box', style='filled', fillcolor='lightyellow')
dot.edge('C_51_illum', 'C_51_illumA', label="Sequence to remove incorrectly hanging red/yellow flag \n 1. Remove red/yellow flag that does not \n intersect with Green Coast flag by bg fill", fontsize='14', fontcolor='black')

dot.node('C_51_illumB', 'Folder: Job_51 Illum \n Red\Yellow flag filled, some unsuitable', shape='box', style='filled', fillcolor='lightyellow')
dot.edge('C_51_illumA', 'C_51_illumB', label=" 2. Fill Red/Yellow flag remants \n where boudning box overlaps wih \n Green Coast flag using HSV thresholds", fontsize='14', fontcolor='black')

dot.node('C_51_illumC', 'Folder: Job_51_Processed \n Green Coast flag only \n Added to "Selected" set \n [178 images 166MB]', shape='box', style='filled', fillcolor='lightgrey')
dot.edge('C_51_illumB', 'C_51_illumC', label="Copy to Selected", fontsize='14', fontcolor='black')

dot.node('C_51_illumD', 'Folder: Job_51_Processed \n  set \n [87 images 81MB]', shape='box', style='filled', fillcolor='lightcoral')
dot.edge('C_51_illumC', 'C_51_illumD', label="Delete by pattern", fontsize='14', fontcolor='black')

dot.node('C_51_illumE', 'Final Job_51_Processed \n  set \n [Final Qty/Ssize]', shape='box', style='filled', fillcolor='lightyellow')
dot.edge('C_51_illumD', 'C_51_illumE', label="Analysis/Reduction", fontsize='14', fontcolor='black')

# Ensure S23FilterMerged, S23Merged, and C_51_illumE are on the same rank (side by side)
with dot.subgraph() as s:
    s.attr(rank='same')
    s.node('S23FilterMerged')
    s.node('S23Merged')
    s.node('C_51_illumE')

# Render the graph
output_file = dot.render('recreated_flowchart', view=True)
print(f"Flowchart saved as: {output_file}")
#S23FilterMerged S23Merged C_51_illumE
#C_51_illumD
```
