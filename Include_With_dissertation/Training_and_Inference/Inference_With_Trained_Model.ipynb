```python
# Inference on a video 

from ultralytics import YOLO
import cv2

# Load the trained model
model = YOLO('best.pt')  

# Path to the video file
#video_path ='https://youtu.be/VjT_x3WQC94'
video_path='D:/FlagDetectionDatasets/A_Videos_Inference_Output/Flag_Composition_For_Inference.mp4'
#video_path = 'D:/Brid_Videos/Home_Videos/Red_yellow_flag_moderate_wind_2.mp4'  
output_path = 'D:/FlagDetectionDatasets/A_Videos_Inference_Output/DetectionOf_Flag3.mp4'  
# Need to add a cv2.imshow() step).
# to display the video while processing 

# Open the video file
cap = cv2.VideoCapture(video_path)
print(video_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define video writer to save output
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Exit the loop if no frame is read

    # Perform inference
    results = model(frame)  # Inference on the current frame

    # Visualize the results
    annotated_frame = results[0].plot()  # Draw boxes and labels on the frame

    # Write the annotated frame to the output video
    out.write(annotated_frame)

    # Optional: Display the video in real time
    cv2.imshow('Video', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to quit
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```
