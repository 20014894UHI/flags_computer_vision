```python
# Synchronise folders - for use after recreatign augmented sets after issues 

import os

import os
import shutil

def synchronize_and_check_folders(main_folder, second_folder):
    """
    Synchronizes two folders by ensuring the text file stems in the main folder
    correspond to files in the second folder. If both folders have matching
    filenames (based on stems), the main folder is deleted.

    Args:
        main_folder (str): Path to the main folder containing text files.
        second_folder (str): Path to the second folder to synchronize with the main folder.
    """
    # Get all text files in the main folder
    main_text_files = [f for f in os.listdir(main_folder) if f.endswith(".txt")]

    # Extract filename stems from main folder text files
    valid_stems = set()
    for text_file in main_text_files:
        if "_scaled_" in text_file:
            stem = text_file.split("_scaled_")[0]
            valid_stems.add(stem)

    # Synchronize second folder
    deleted_files = []
    for file_name in os.listdir(second_folder):
        file_path = os.path.join(second_folder, file_name)
        # Check if the stem of the file matches any stem in the main folder
        if "_scaled_" in file_name:
            stem = file_name.split("_scaled_")[0]
            if stem not in valid_stems:
                # Delete the file
                print(f"Deleting {file_path}")
                os.remove(file_path)
                deleted_files.append(file_name)

    # Check if the two folders have matching stems
    second_folder_stems = {
        file_name.split("_scaled_")[0]
        for file_name in os.listdir(second_folder)
        if "_scaled_" in file_name
    }

    unmatched_stems = valid_stems.symmetric_difference(second_folder_stems)

    # If both folders have matching stems, delete the main folder
    if not unmatched_stems:
        print("All stems match. Deleting main folder...")
        # shutil.rmtree(main_folder)
        print(f"Deleted main folder: {main_folder}")
    else:
        print("Folders do not have matching stems. Not deleting the main folder.")
        print(f"Unmatched stems: {unmatched_stems}")



def synchronize_foldersReplaced(main_folder, second_folder):
    """
    Synchronizes two folders by deleting files in the second folder that do not have corresponding text files in the main folder.

    Args:
        main_folder (str): Path to the main folder containing text files.
        second_folder (str): Path to the second folder to synchronize with the main folder.
    """
    # Get all text files in the main folder
    main_text_files = [f for f in os.listdir(main_folder) if f.endswith(".txt")]

    # Extract filename stems from main folder text files
    valid_stems = set()
    for text_file in main_text_files:
        if "_scaled_" in text_file:
            stem = text_file.split("_scaled_")[0]
            valid_stems.add(stem)

    # Iterate through files in the second folder
    for file_name in os.listdir(second_folder):
        file_path = os.path.join(second_folder, file_name)
        # Check if the stem of the file matches any stem in the main folder
        if "_scaled_" in file_name:
            stem = file_name.split("_scaled_")[0]
            if stem not in valid_stems:
                # Delete the text file or image
                print(f"Deleting {file_path}")
                os.remove(file_path)

    print("Synchronization complete.")


# Careful ! DELETES THE MAIN FOLDER AFTER SYNCH  !!! 
# Delete items from the 2nd folder that are not in the main folder based on the stem - the part before "_scaled_"
# Deletes the main folder then 
main_folder   = "D:/FlagDetectionDatasets/ExportedDatasetsSelected/Job_115_Aug"
second_folder = "D:/FlagDetectionDatasets/ExportedDatasetsSelected/Job_115_Aug_T"
#synchronize_folders(main_folder, second_folder)
synchronize_and_check_folders(main_folder, second_folder)
# To delete the folder Uncomment the line   # shutil.rmtree(main_folder) to restore deleting the folder 



```
